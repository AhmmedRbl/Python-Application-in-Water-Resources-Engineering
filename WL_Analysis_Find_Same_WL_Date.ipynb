```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wl_df = pd.read_csv(r'E:\MSc RBE\Future Image Generation\WL\SW_91_9_L_Baruria_Transit.csv')
wl_df.head()
```

```python
# Filter for months December (12), January (1), and February (2)
winter_df = wl_df[wl_df['Month'].isin([12, 1, 2])]

# Group by year and find the minimum WL_mMSL for each year
min_wl_per_year = winter_df.groupby('Year')['WL_mMSL'].min()

print(min_wl_per_year)
```

```python
# Create a DataFrame with all dates where WL_mMSL is between 1.75 and 2.25, grouped by year

wl_range_df = winter_df[(winter_df['WL_mMSL'] >= 1.75) & (winter_df['WL_mMSL'] <= 2.25)]
dates_in_range_df = wl_range_df.groupby('Year').agg({'Date.1': list}).reset_index()
dates_in_range_df.head(25)
```

```python
wl_range_df['Date_only'] = wl_range_df['Date.1'].str.split(' ').str[0]
dates_only_in_range_df = wl_range_df.groupby('Year').agg({'Date_only': lambda x: sorted(set(x))}).reset_index()
dates_only_in_range_df.head(25)
```
